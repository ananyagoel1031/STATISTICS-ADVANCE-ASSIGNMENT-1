```python
#Q9

import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from a normal distribution
mean = 50
std_dev = 5
data = np.random.normal(loc=mean, scale=std_dev, size=1000)

# Compute the mean and standard deviation of the generated data
computed_mean = np.mean(data)
computed_std_dev = np.std(data)

# Print the computed mean and standard deviation
print(f"Computed Mean: {computed_mean:.2f}")
print(f"Computed Standard Deviation: {computed_std_dev:.2f}")

# Plot a histogram to visualize the distribution
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Generated Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.axvline(computed_mean, color='red', linestyle='dashed', linewidth=1, 
label=f'Mean: {computed_mean:.2f}')
plt.legend()
plt.show()
```

```python
#Q10
import numpy as np
from scipy import stats

# Example sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]


# Calculate the mean and standard error of the mean
mean_sales = np.mean(daily_sales)
sem_sales = stats.sem(daily_sales)

# Define the confidence level (e.g., 95%)
confidence_level = 0.95
degrees_of_freedom = len(daily_sales) - 1

# Calculate the confidence interval
confidence_interval = stats.t.interval(
    confidence_level, 
    degrees_of_freedom, 
    loc=mean_sales, 
    scale=sem_sales
)

print(f"Mean Sales: {mean_sales}")
print(f"95% Confidence Interval: {confidence_interval}")
```
